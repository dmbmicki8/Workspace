## UN Data API Exercise

#### In this exercise, you'll redo the data gathering phase of the UNData Exploration project by using APIs instead of downloading csv files. You'll make use of the [World Bank Indicators API](https://datahelpdesk.worldbank.org/knowledgebase/articles/889392-about-the-indicators-api-documentation). Note that this API does not require an API key. Before attempting the exercise, it would be a good idea to skim through the Documentation page and to check out the [Basic Call Structure article](https://datahelpdesk.worldbank.org/knowledgebase/articles/898581).

```python
import requests
import pandas as pd
```

##### 1. Use the API to get all available data for the _GDP per capita, PPP (constant 2017 international $)_ indicator. Hint: this indicator has code "NY.GDP.PCAP.PP.KD". Adjust the query parameters so that you can retrieve all available rows. Convert the results to a DataFrame.


```python
gdp_endpoint = 'https://api.worldbank.org/V2/country/all/indicator/NY.GDP.PCAP.PP.KD'

# Set parameters
params = {
    'format': 'json',
    'per_page': 20000
}

# Make the request
gdp_response = requests.get(gdp_endpoint, params=params)

# Check it worked
print(gdp_response.status_code)
```

```python
# Parse the JSON response
res = gdp_response.json()

# Check the metadata (total records, pages, etc.)
print(res[0])
```

```python
# The actual data is in item [1]
gdp_df = pd.DataFrame(res[1])

# Preview the first few rows
gdp_df.head()
```

```python
# Extract just the useful values from the nested dictionaries
gdp_df['indicator'] = gdp_df['indicator'].apply(lambda x: x['value'])
gdp_df['country'] = gdp_df['country'].apply(lambda x: x['value'])

# Preview the cleaned DataFrame
gdp_df.head()
```

##### 2. Now, use the API to get all available data for _Life expectancy at birth, total (years)_. This indicator has code "SP.DYN.LE00.IN". Again, convert the results to a DataFrame.


```python
# Step 2 - Life Expectancy
life_endpoint = 'https://api.worldbank.org/V2/country/all/indicator/SP.DYN.LE00.IN'

params = {
    'format': 'json',
    'per_page': 20000
}

life_response = requests.get(life_endpoint, params=params)
life_res = life_response.json()

# Check metadata
print(life_res[0])
```

```python
# Convert to DataFrame
life_df = pd.DataFrame(life_res[1])

# Extract values from nested dictionaries
life_df['indicator'] = life_df['indicator'].apply(lambda x: x['value'])
life_df['country'] = life_df['country'].apply(lambda x: x['value'])

# Preview
life_df.head()
```

##### 3. Merge the two results DataFrames together. You may want to rename or drop columns prior to merging.


```python
# Keep only the columns we need and rename 'value'
gdp_clean = gdp_df[['country', 'countryiso3code', 'date', 'value']].rename(columns={'value': 'gdp_per_capita'})

life_clean = life_df[['country', 'countryiso3code', 'date', 'value']].rename(columns={'value': 'life_expectancy'})

# Merge on country and date
merged_df = pd.merge(gdp_clean, life_clean, on=['country', 'countryiso3code', 'date'])

# Preview
merged_df.head()
```

##### 4. You can also get more information about the available countries (region, capital city, income level classification, etc.) by using the [Country API](https://datahelpdesk.worldbank.org/knowledgebase/articles/898590-country-api-queries). Use this API to pull in all available data. Merge this with your other datasets. Use this to now remove the rows that correspond to regions and not countries.


```python
# Step 4 - Country API
country_endpoint = 'https://api.worldbank.org/V2/country/all'

params = {
    'format': 'json',
    'per_page': 500  # There aren't that many countries so 500 is plenty
}

country_response = requests.get(country_endpoint, params=params)
country_res = country_response.json()

# Check metadata
print(country_res[0])
```

```python
# Convert to DataFrame
country_df = pd.DataFrame(country_res[1])

# Preview
country_df.head()
```

```python
# Extract values from nested dictionaries
country_df['region'] = country_df['region'].apply(lambda x: x['value'])
country_df['adminregion'] = country_df['adminregion'].apply(lambda x: x['value'])
country_df['incomeLevel'] = country_df['incomeLevel'].apply(lambda x: x['value'])
country_df['lendingType'] = country_df['lendingType'].apply(lambda x: x['value'])

# Preview
country_df.head()
```

```python
print(country_df.columns.tolist())
```

```python
country_clean = country_df[['id', 'name', 'region', 'incomeLevel', 'capitalCity']]

# Preview
country_clean.head()
```

```python
# Merge country info with our merged_df
# 'id' in country_clean matches 'countryiso3code' in merged_df
final_df = pd.merge(merged_df, country_clean, left_on='countryiso3code', right_on='id')

# Now filter out regions - actual countries have a real region, not 'Aggregates'
final_df = final_df[final_df['region'] != 'Aggregates']

# Drop the duplicate 'id' column since we already have 'countryiso3code'
final_df = final_df.drop(columns='id')

# Preview
final_df.head()
```

```python
print(final_df.shape)
print(final_df['country'].nunique())
```

##### 5. If you haven't already done so and you would like to get some additional practice using loops, use the page parameter in order to pull all records. Do not change the value of the per_page parameter. You will likely need to utilize a loop of some kind in order to pull all records.


```python
# Use default per_page and loop through all pages
all_data = []
page = 1

while True:
    params = {
        'format': 'json',
        'per_page': 50,  # default, don't change it
        'page': page
    }
    
    response = requests.get(gdp_endpoint, params=params)
    res = response.json()
    
    # Add this page's data to list
    all_data.extend(res[1])
    
    # Stop going through all pages
    if page >= res[0]['pages']:
        break
    
    page += 1

print(f'Total records collected: {len(all_data)}')
```

```python
all_data = []
page = 1

while True:
    params = {
        'format': 'json',
        'per_page': 50,
        'page': page
    }
    
    response = requests.get(gdp_endpoint, params=params)
    
    # Check status before trying to parse
    print(f'Page {page} - Status: {response.status_code}')
    
    if response.status_code != 200:
        print('Error, stopping loop')
        break
        
    res = response.json()
    all_data.extend(res[1])
    
    if page >= res[0]['pages']:
        break
    
    page += 1

print(f'Total records collected: {len(all_data)}')
```
