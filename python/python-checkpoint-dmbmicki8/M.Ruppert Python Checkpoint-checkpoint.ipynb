#### 8d.Use the **IQR method** to determine the upper and lower cutoffs for outliers.

```python
Q3 = movies2003['duration'].quantile(0.75)
Q1 = movies2003['duration'].quantile(0.25)
IQR = Q3 - Q1

print(IQR)
```

```python
lower = Q1-(1.5*IQR)
print(lower)
higher = Q3+(1.5*IQR)
print(higher)
```

```python
movies2003[(movies2003['duration'] < lower) | (movies2003['duration'] > higher)]
```

```python
movies2003[(movies2003['duration'] < lower) | (movies2003['duration'] >higher)].shape[0] / movies2003.shape[0] * 100
```

### Add a new column called **`length_category`** to `movies_df` that categorizes each movie as one of the following:

| Category  | Duration               |
|-----------|------------------------|
| `short`   | 30 minutes or less     |
| `average` | Between 31–120 minutes |
| `long`    | More than 120 minutes  |


**Bonus:** Once the column is created, create a **pie chart** showing the distribution of movie length categories.


```python
movies_df = disney[disney.type == 'Movie'].copy()
movies_df['duration'] = movies_df['duration'].str.replace('min', '').astype(int)
```

```python
import numpy as np
```

```python
conditions = [
    movies_df['duration'] <= 30,
    (movies_df['duration'] >= 31) & (movies_df['duration'] <= 120),
    movies_df['duration'] > 120
]
labels = ['short', 'average', 'long']
movies_df['length_category'] = np.select(conditions, labels, default='unknown')
```

```python
movies_df[['duration', 'length_category']].head(10)
```

```python
movies_df['length_category'].value_counts().plot.pie()
```
