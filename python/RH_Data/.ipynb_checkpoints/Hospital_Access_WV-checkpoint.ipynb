```python
import geopandas as gpd
import pandas as pd
import folium
from pygris import counties
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import time
```

```python
# Load county shapefile and filter to WV only
counties_us = gpd.read_file(r"C:\Users\Micki\workspace\python\RH_Data\tl_2025_us_county.shp")
wv_counties = counties_us[counties_us['STATEFP'] == '54'].copy()
wv_counties['NAME'] = wv_counties['NAME'].str.title().str.strip()
```

```python
# Load hospitals
hospitals = pd.read_csv(r"C:\Users\Micki\workspace\python\RH_Data\WV_Hospitals.csv")
hospitals['County/Parish'] = hospitals['County/Parish'].str.title().str.strip()
hospitals['County/Parish'] = hospitals['County/Parish'].str.replace('Mc ', 'Mc', regex=False)
```

```python
# Load HRSA clean data
hrsa_clean = pd.read_csv(r"C:\Users\Micki\workspace\python\RH_Data\HRSA_WV_Clean.csv")
hrsa_clean['County Name'] = hrsa_clean['County Name'].str.replace(' County', '').str.strip()
```

```python
# Load MUA clean data
mua = pd.read_csv(r"C:\Users\Micki\workspace\python\RH_Data\MUA_WV_Clean.csv")
mua['County'] = mua['County'].str.replace(' County, WV', '').str.strip()
mua['County'] = mua['County'].str.replace(' County', '').str.strip()
```

```python
# Geocode hospitals
geolocator = Nominatim(user_agent="rural_health_wv")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

hospitals['full_address'] = hospitals['Address'] + ', ' + hospitals['City/Town'] + ', WV'
hospitals['location'] = hospitals['full_address'].apply(geocode)
hospitals['lat'] = hospitals['location'].apply(lambda x: x.latitude if x else None)
hospitals['lng'] = hospitals['location'].apply(lambda x: x.longitude if x else None)

# Fix nulls using city name
for idx in hospitals[hospitals['lat'].isna()].index:
    city = hospitals.loc[idx, 'City/Town']
    location = geocode(city + ', WV')
    if location:
        hospitals.loc[idx, 'lat'] = location.latitude
        hospitals.loc[idx, 'lng'] = location.longitude
    time.sleep(1)

print("Hospital nulls:", hospitals['lat'].isna().sum())
```

```python
# Build counts and merge everything into wv_map
hospital_counts = hospitals.groupby('County/Parish').size().reset_index(name='hospital_count')

hrsa_primary = hrsa_clean[hrsa_clean['Discipline'] == 'Primary Care']
hrsa_counts = hrsa_primary.groupby('County Name').size().reset_index(name='hpsa_count')

mua_counts = mua.groupby('County').size().reset_index(name='mua_count')

wv_map = wv_counties.merge(hospital_counts, left_on='NAME', right_on='County/Parish', how='left')
wv_map = wv_map.merge(hrsa_counts, left_on='NAME', right_on='County Name', how='left')
wv_map = wv_map.merge(mua_counts, left_on='NAME', right_on='County', how='left')

wv_map['hospital_count'] = wv_map['hospital_count'].fillna(0)
wv_map['hpsa_count'] = wv_map['hpsa_count'].fillna(0)
wv_map['mua_count'] = wv_map['mua_count'].fillna(0)

print(wv_map[['NAME', 'hospital_count', 'hpsa_count', 'mua_count']].head(10))
```

```python
# Calculate center from your actual WV data
wv_center_lat = wv_counties.geometry.centroid.y.mean()
wv_center_lng = wv_counties.geometry.centroid.x.mean()

print(wv_center_lat, wv_center_lng)
```

```python
# Aggregate HPSA designation names per county
hpsa_names = hrsa_primary.groupby('County Name')['HPSA Name'].apply(lambda x: ', '.join(x)).reset_index()
hpsa_names.columns = ['County Name', 'hpsa_names']

# Aggregate MUA designation names per county
mua_names = mua.groupby('County')['Service Area Name'].apply(lambda x: ', '.join(x)).reset_index()
mua_names.columns = ['County', 'mua_names']

# Merge into wv_map
wv_map = wv_map.merge(hpsa_names, left_on='NAME', right_on='County Name', how='left')
wv_map = wv_map.merge(mua_names, left_on='NAME', right_on='County', how='left')

# Fill nulls
wv_map['hpsa_names'] = wv_map['hpsa_names'].fillna('None')
wv_map['mua_names'] = wv_map['mua_names'].fillna('None')

print(wv_map[['NAME', 'hpsa_names', 'mua_names']].head(10))
```

```python
wv_center = [wv_center_lat, wv_center_lng]
m = folium.Map(location=wv_center, zoom_start=8)

# Choropleth using MUA designations
choropleth = folium.Choropleth(
    geo_data=wv_map,
    name='Medically Underserved Areas',
    data=wv_map,
    columns=['NAME', 'mua_count'],
    key_on='feature.properties.NAME',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Medically Underserved Area Designations'
).add_to(m)

# Add hover tooltips showing county name, MUA count, and HPSA count
choropleth.geojson.add_child(
    folium.features.GeoJsonTooltip(
        fields=['NAME', 'mua_names', 'hpsa_names', 'hospital_count'],
        aliases=['County:', 'MUA Designations:', 'HPSA Designations:', 'Hospitals:'],
        style='font-size: 13px;'
    )
)

# Hospital points (blue)
for idx, row in hospitals.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lng']],
        radius=4,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.7,
        popup=folium.Popup(
            f"<b>{row['Facility Name']}</b><br>{row['Hospital Type']}<br>{row['City/Town']}, WV",
            max_width=200
        )
    ).add_to(m)

# Legend
legend_html = '''
<div style="position: fixed; bottom: 50px; left: 50px; z-index: 1000; 
     background-color: white; padding: 10px; border-radius: 5px;
     border: 2px solid grey; font-size: 14px;">
     <b>WV Rural Health Deserts</b><br>
     <i class="fa fa-circle" style="color:blue"></i> Hospital Location<br>
     <span style="color:red">&#9632;</span> Medically Underserved Area Designations<br>
     <i>Hover over county for details</i>
</div>
'''
m.get_root().html.add_child(folium.Element(legend_html))

# Save and display
m.save(r"C:\Users\Micki\workspace\python\RH_Data\rural_health_map.html")
m
```
